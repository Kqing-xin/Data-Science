**获取所有的股票60日涨跌幅数据**

```python
import tushare as ts
import pandas as pd
import datetime
token = ""
pro = ts.pro_api(token)  #初始化pro接口

# 计算N日涨跌幅数据
def stock_N_pct_change(stockCode=None,stockTimeS=None,stockTimeE=None, N=None):         
    df_stockload = pro.daily(ts_code=stockCode,start_date=stockTimeS,end_date=stockTimeE)
    df_stockload.trade_date = pd.to_datetime(df_stockload.trade_date)
    df_stockload.set_index(df_stockload.trade_date,drop=True,inplace=True)
    df_stockload.sort_index(inplace=True)
    df_stockload[f'{N}day_pct_change'] = df_stockload.close.pct_change(N)  # 与前N日进行比较
    df_stockload[f'{N}day_pct_change'].fillna(0,inplace=True)   # 填充空值
    return df_stockload

# 读取A股代码名称
stock_codes = pd.read_csv('A股代码及名称.csv')
stock_codes = stock_codes
stock_codes.set_index("证券代码",inplace=True,drop=True)
stock_codes.drop(columns="Unnamed: 0",inplace=True)
stock_codes

# 定义时间周期
now_time =datetime.datetime.now()#获取当前时间
yes_time = (now_time - datetime.timedelta(365)) #一年前时间
now_time = now_time.strftime('%Y%m%d')#获取当前时间
yes_time = yes_time.strftime('%Y%m%d')
print(now_time,yes_time)


# 定义周期及计算
import time
x = 0
all_stock_N_pct_change = pd.DataFrame()
N = 60   # 周期选择   60/120/250
for index,row in stock_codes.iterrows():
    code = index
    name = row.证券简称

    df_stockload = stock_N_pct_change(stockCode=code,stockTimeS=yes_time,stockTimeE=now_time,N=N)
    N_pct_change = df_stockload[f"{N}day_pct_change"].rename(name)
    all_stock_N_pct_change = pd.concat([all_stock_N_pct_change,N_pct_change],join='outer',axis=1)
    time.sleep(0.005)
    x = x+1
    print(f"到第{x}个,共4424个")
print("遍历完成")
```

```python
all_stock_N_pct_change.to_csv('all_stock_N_pct_change.csv',enconding="gbk")
```
